```python
# @title Hayvan Sınıflandırıcı Model Eğitimi

!pip install torch torchvision tqdm matplotlib scikit-learn

import os
import time
import copy
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from torchvision import transforms, datasets, models
from torch.utils.data import DataLoader, random_split
from google.colab import drive, files
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Google Drive'a bağlan
drive.mount('/content/drive', force_remount=True)

# Çalışma dizinini oluştur
!mkdir -p /content/animal_classifier/data
!mkdir -p /content/animal_classifier/models

# Veri setini indir (Kaggle'dan indirdiğiniz veri setini Google Drive'a yüklemeniz gerekecek)
# Drive'dan veri setini kopyala (eğer varsa)
!cp -r /content/drive/MyDrive/animals-10.zip /content/animal_classifier/data/ 2>/dev/null || echo "Veri seti Drive'da bulunamadı."

# Veri seti Drive'da yoksa Kaggle'dan indir
!pip install -q kaggle
!mkdir -p ~/.kaggle

# Kaggle API anahtarınızı burada kullanmanız gerekecek
# @param {type:"string"}
kaggle_username = "muhammedseyrek" # @param {type:"string"}
# @param {type:"string"}
kaggle_key = "" # @param {type:"string"}

!echo "{\"username\":\"$kaggle_username\",\"key\":\"$kaggle_key\"}" > ~/.kaggle/kaggle.json
!chmod 600 ~/.kaggle/kaggle.json

# Kaggle'dan Animals-10 veri setini indir (eğer henüz indirilmediyse)
!ls -la /content/animal_classifier/data/
if not os.path.exists('/content/animal_classifier/data/animals-10.zip'):
  !kaggle datasets download -d alessiocorrado99/animals10
  !mv animals10.zip /content/animal_classifier/data/animals-10.zip

# Veri setini çıkart
!unzip -q /content/animal_classifier/data/animals-10.zip -d /content/animal_classifier/data/animals-10

# İtalyanca-İngilizce çeviri sözlüğü
translation = {
    'cane': 'dog',
    'cavallo': 'horse',
    'elefante': 'elephant',
    'farfalla': 'butterfly',
    'gallina': 'chicken',
    'gatto': 'cat',
    'mucca': 'cow',
    'pecora': 'sheep',
    'ragno': 'spider',
    'scoiattolo': 'squirrel'
}

def get_english_class_name(italian_name):
    """İtalyanca sınıf adlarını İngilizce'ye çevirir."""
    return translation.get(italian_name, italian_name)

class AnimalClassifier(nn.Module):
    """Transfer öğrenme kullanarak hayvan sınıflandırıcı modeli."""

    def __init__(self, num_classes, model_name='resnet18', pretrained=True):
        """
        Parametreler:
            num_classes (int): Sınıflandırılacak hayvan sınıfı sayısı.
            model_name (str): Kullanılacak temel model adı.
            pretrained (bool): Önceden eğitilmiş ağırlıkları kullanıp kullanmama.
        """
        super(AnimalClassifier, self).__init__()

        self.model_name = model_name

        # Model seçimi
        if model_name == 'resnet18':
            self.base_model = models.resnet18(weights='IMAGENET1K_V1' if pretrained else None)
            num_features = self.base_model.fc.in_features
            self.base_model.fc = nn.Identity()  # Son katmanı kaldır

        elif model_name == 'mobilenet_v2':
            self.base_model = models.mobilenet_v2(weights='IMAGENET1K_V1' if pretrained else None)
            num_features = self.base_model.classifier[1].in_features
            self.base_model.classifier = nn.Identity()  # Son katmanı kaldır

        else:
            raise ValueError(f"Desteklenmeyen model: {model_name}")

        # Yeni sınıflandırıcı katmanı
        self.classifier = nn.Sequential(
            nn.Dropout(0.2),
            nn.Linear(num_features, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        """İleri geçiş."""
        features = self.base_model(x)
        return self.classifier(features)

    def freeze_base_model(self):
        """Transfer öğrenme için temel modelin parametrelerini dondur."""
        for param in self.base_model.parameters():
            param.requires_grad = False

    def unfreeze_base_model(self):
        """Fine-tuning için temel modelin parametrelerini çöz."""
        for param in self.base_model.parameters():
            param.requires_grad = True

    def save_model(self, path):
        """Modeli kaydet."""
        torch.save({
            'model_state_dict': self.state_dict(),
            'model_name': self.model_name
        }, path)

    @classmethod
    def load_model(cls, path, num_classes):
        """Kaydedilmiş modeli yükle."""
        checkpoint = torch.load(path, map_location=torch.device('cpu'))
        model = cls(num_classes=num_classes, model_name=checkpoint['model_name'], pretrained=False)
        model.load_state_dict(checkpoint['model_state_dict'])
        model.eval()
        return model

class AnimalDataProcessor:
    """Sınıf, hayvan görüntüleri veri setini işlemek, dönüştürmek ve yüklemek için kullanılır."""

    def __init__(self, data_dir, img_size=224, batch_size=32, val_split=0.2, test_split=0.1):
        """
        Parametreler:
            data_dir (str): Veri setinin bulunduğu dizin.
            img_size (int): Görüntülerin yeniden boyutlandırılacağı piksel boyutu.
            batch_size (int): Eğitim ve değerlendirme için batch boyutu.
            val_split (float): Doğrulama seti için ayrılacak veri oranı (0-1 arası).
            test_split (float): Test seti için ayrılacak veri oranı (0-1 arası).
        """
        self.data_dir = data_dir
        self.img_size = img_size
        self.batch_size = batch_size
        self.val_split = val_split
        self.test_split = test_split

        # Eğitim için dönüşüm işlemleri tanımlanıyor (veri çoğaltma dahil)
        self.train_transforms = transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(20),
            transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            )
        ])

        # Test için dönüşüm işlemleri tanımlanıyor (sadece normalizasyon)
        self.test_transforms = transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            )
        ])

        # Tahmin için dönüşüm işlemleri (uygulama için)
        self.predict_transforms = self.test_transforms

        self.class_names = []
        self.english_class_names = []

    def load_data(self):
        """Veri setini yükler, eğitim, doğrulama ve test setlerini hazırlar."""
        # Animals-10 veri setini yükle
        full_dataset = datasets.ImageFolder(self.data_dir, transform=self.train_transforms)

        # Sınıf isimlerini kaydet
        self.class_names = full_dataset.classes
        self.class_to_idx = full_dataset.class_to_idx

        # İtalyanca sınıf isimlerini İngilizce'ye çevir
        self.english_class_names = [get_english_class_name(class_name) for class_name in self.class_names]

        # Veri setini eğitim, doğrulama ve test setlerine böl
        dataset_size = len(full_dataset)
        test_size = int(dataset_size * self.test_split)
        val_size = int(dataset_size * self.val_split)
        train_size = dataset_size - val_size - test_size

        train_dataset, val_dataset, test_dataset = random_split(
            full_dataset, [train_size, val_size, test_size],
            generator=torch.Generator().manual_seed(42)
        )

        # Dönüşümleri uygula (eğitim için data augmentation, test ve val için sadece normalizasyon)
        train_dataset.dataset.transform = self.train_transforms
        val_dataset.dataset.transform = self.test_transforms
        test_dataset.dataset.transform = self.test_transforms

        # DataLoader'ları oluştur
        train_loader = DataLoader(
            train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=2
        )

        val_loader = DataLoader(
            val_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2
        )

        test_loader = DataLoader(
            test_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2
        )

        print(f"Toplam sınıf sayısı: {len(self.class_names)}")
        print(f"Eğitim örnekleri: {train_size}")
        print(f"Doğrulama örnekleri: {val_size}")
        print(f"Test örnekleri: {test_size}")

        return train_loader, val_loader, test_loader

    def visualize_samples(self, num_samples=5):
        """Veri setinden rastgele örnekleri görselleştirme."""
        # Görselleştirme için dönüşümlerden ToTensor ve Normalize çıkartılmalı
        vis_transforms = transforms.Compose([
            transforms.Resize((self.img_size, self.img_size)),
            transforms.ToTensor(),
        ])

        dataset = datasets.ImageFolder(self.data_dir, transform=vis_transforms)

        # Rastgele indeksler seç
        indices = torch.randperm(len(dataset))[:num_samples]

        plt.figure(figsize=(15, 3))
        for i, idx in enumerate(indices):
            img, label = dataset[idx]
            # Tensörü numpy dizisine dönüştür
            img = img.permute(1, 2, 0).numpy()  # [C, H, W] -> [H, W, C]

            plt.subplot(1, num_samples, i+1)
            plt.imshow(img)
            plt.title(f"{self.class_names[label]} ({self.english_class_names[label]})")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

def train_model(model, dataloaders, criterion, optimizer, scheduler, device,
                num_epochs=25, save_path='./models/best_model.pth'):
    """Model eğitimi için fonksiyon."""
    since = time.time()

    # En iyi modelin kopyasını tutacağız
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    # Eğitim metrikleri
    history = {
        'train_loss': [],
        'val_loss': [],
        'train_acc': [],
        'val_acc': []
    }

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        # Her epoch'ta hem eğitim hem de validasyon aşamaları var
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Eğitim modu (dropout aktif, batch norm güncellenir)
            else:
                model.eval()   # Değerlendirme modu

            running_loss = 0.0
            running_corrects = 0

            # Veri kümesi üzerinden döngü
            pbar = tqdm(dataloaders[phase], desc=f'{phase} iteration')
            for inputs, labels in pbar:
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Parametrelerin gradyanlarını sıfırla
                optimizer.zero_grad()

                # İleri geçiş
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # Eğitim modunda geriye yayılım + optimizasyon
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # İstatistikleri topla
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            if phase == 'train' and scheduler is not None:
                scheduler.step()

            # Epoch istatistiklerini hesapla
            epoch_loss = running_loss / len(dataloaders[phase].dataset)
            epoch_acc = running_corrects.double() / len(dataloaders[phase].dataset)

            # Metrikleri kaydet
            if phase == 'train':
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc.item())
            else:
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc.item())

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            # En iyi modeli kaydet
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())

                # Modeli kaydet
                model.save_model(save_path)
                print(f'En iyi model kaydedildi: {save_path}')

        print()

    time_elapsed = time.time() - since
    print(f'Eğitim tamamlandı: {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'En iyi validasyon Acc: {best_acc:.4f}')

    # En iyi model ağırlıklarını yükle
    model.load_state_dict(best_model_wts)
    return model, history

def plot_training_history(history):
    """Eğitim metriklerini görselleştir."""
    epochs = range(1, len(history['train_loss']) + 1)

    plt.figure(figsize=(12, 5))

    # Kayıp grafiği
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['train_loss'], 'bo-', label='Eğitim Kaybı')
    plt.plot(epochs, history['val_loss'], 'ro-', label='Validasyon Kaybı')
    plt.title('Eğitim ve Validasyon Kaybı')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Doğruluk grafiği
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['train_acc'], 'bo-', label='Eğitim Doğruluğu')
    plt.plot(epochs, history['val_acc'], 'ro-', label='Validasyon Doğruluğu')
    plt.title('Eğitim ve Validasyon Doğruluğu')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.savefig('/content/animal_classifier/models/training_history.png')
    plt.show()

def evaluate_model(model, test_loader, device, class_names):
    """Test veri seti üzerinde modeli değerlendirir."""
    model.eval()  # Değerlendirme modu

    all_preds = []
    all_labels = []
    correct = 0
    total = 0

    # Gradyanları devre dışı bırak
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Test"):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            # İstatistikleri topla
            total += labels.size(0)
            correct += (preds == labels).sum().item()

            # Tahminleri ve etiketleri kaydet
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Doğruluğu hesapla
    accuracy = correct / total
    print(f'Test Doğruluğu: {accuracy:.4f}')

    # Karışıklık matrisini hesapla
    cm = confusion_matrix(all_labels, all_preds)

    # Karışıklık matrisini görselleştir
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Tahmin Edilen')
    plt.ylabel('Gerçek')
    plt.title('Karışıklık Matrisi')
    plt.tight_layout()
    plt.savefig('/content/animal_classifier/models/confusion_matrix.png')
    plt.show()

    # Sınıflandırma raporu oluştur
    report = classification_report(all_labels, all_preds, target_names=class_names, digits=4)
    print("Sınıflandırma Raporu:")
    print(report)

    # Raporu dosyaya kaydet
    with open('/content/animal_classifier/models/classification_report.txt', 'w') as f:
        f.write(report)

    return accuracy, all_preds, all_labels

# Ana program
def main():
    # Veri işleyiciyi oluştur
    data_processor = AnimalDataProcessor(
        data_dir="/content/animal_classifier/data/animals-10/raw-img",
        img_size=224,
        batch_size=32
    )

    # Veri yükleyicileri oluştur
    train_loader, val_loader, test_loader = data_processor.load_data()

    # DataLoaders sözlüğü
    dataloaders = {
        'train': train_loader,
        'val': val_loader
    }

    # Cihazı seç
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Eğitim için kullanılan cihaz: {device}")

    # Sınıf sayısını al
    num_classes = len(data_processor.class_names)
    print(f"Sınıf sayısı: {num_classes}")

    # Örnek görüntüleri göster
    data_processor.visualize_samples(num_samples=5)

    # Modeli oluştur (daha küçük bir model kullanalım)
    model = AnimalClassifier(num_classes=num_classes, model_name='resnet18')

    # İlk aşamada temel modeli dondur (sadece sınıflandırıcı katmanı eğit)
    model.freeze_base_model()

    # Modeli cihaza taşı
    model = model.to(device)

    # Kayıp fonksiyonu ve optimizasyon algoritması
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)

    # Öğrenme oranı zamanlayıcısı
    scheduler = lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)

    # Modeli eğit (sadece sınıflandırıcı katmanlar için)
    print("İlk aşama: Sadece sınıflandırıcı katmanları eğitme")
    model, history1 = train_model(
        model, dataloaders, criterion, optimizer, scheduler,
        device, num_epochs=5, save_path='/content/animal_classifier/models/model_stage1.pth'
    )

    # İkinci aşama: Temel modeli de eğit (fine-tuning)
    print("\nİkinci aşama: Tüm modeli ince ayarlama")
    model.unfreeze_base_model()  # Temel modelin parametrelerini çöz

    # Tüm parametreler için daha düşük öğrenme oranı
    optimizer = optim.Adam(model.parameters(), lr=0.0001)
    scheduler = lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)

    model, history2 = train_model(
        model, dataloaders, criterion, optimizer, scheduler,
        device, num_epochs=3, save_path='/content/animal_classifier/models/model_final.pth'
    )

    # Eğitim metriklerini birleştir
    history = {
        'train_loss': history1['train_loss'] + history2['train_loss'],
        'val_loss': history1['val_loss'] + history2['val_loss'],
        'train_acc': history1['train_acc'] + history2['train_acc'],
        'val_acc': history1['val_acc'] + history2['val_acc']
    }

    # Eğitim geçmişini görselleştir
    plot_training_history(history)

    # Test setinde değerlendir
    evaluate_model(model, test_loader, device, data_processor.english_class_names)

    # Modeli kaydet
    model.save_model('/content/animal_classifier/models/model_final.pth')

    # Gerekli dosyaları indir
    print("Modeli ve dosyaları indirme...")
    files.download('/content/animal_classifier/models/model_final.pth')
    files.download('/content/animal_classifier/models/training_history.png')
    files.download('/content/animal_classifier/models/confusion_matrix.png')
    files.download('/content/animal_classifier/models/classification_report.txt')

    # Meta verileri kaydet
    class_info = {
        'class_names': data_processor.class_names,
        'english_class_names': data_processor.english_class_names,
        'class_to_idx': data_processor.class_to_idx
    }

    import json
    with open('/content/animal_classifier/models/class_info.json', 'w') as f:
        json.dump(class_info, f)

    files.download('/content/animal_classifier/models/class_info.json')

    # Ayrıca model örnek görüntülerini kaydet ve indir
    import random

    # Önceden tanımlanmış çeviri sözlüğü
    sample_images_dir = '/content/animal_classifier/sample_images'
    os.makedirs(sample_images_dir, exist_ok=True)

    # Her sınıftan bir örnek al
    for class_name in data_processor.class_names:
        class_dir = os.path.join('/content/animal_classifier/data/animals-10/raw-img', class_name)
        images = os.listdir(class_dir)

        # Rastgele bir görüntü seç
        sample_image = random.choice(images)
        sample_path = os.path.join(class_dir, sample_image)

        # Yeni dosya adı (İngilizce sınıf adı ile)
        english_name = get_english_class_name(class_name)
        dest_path = os.path.join(sample_images_dir, f'{english_name}.jpg')

        # Dosyayı kopyala
        import shutil
        shutil.copy(sample_path, dest_path)

    # Tüm örnek görüntüleri zip'le ve indir
    !zip -r /content/animal_classifier/sample_images.zip /content/animal_classifier/sample_images
    files.download('/content/animal_classifier/sample_images.zip')

# Programı çalıştır
main()
```
